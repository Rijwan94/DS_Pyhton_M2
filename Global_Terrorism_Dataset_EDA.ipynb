## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Data Wrangling Code - EDA - Global Terrorism Dataset
class TerrorismWrangling:
    def __init__(self, gtd_df):
        """Initialize with dataframe."""
        try:
            self.df = gtd_df.copy()
            # Convert columns to appropriate data types
            self.df['Year'] = self.df['Year'].astype(int)

            # Handle non-numeric values in 'Ransom' before conversion
            self.df['Ransom'] = self.df['Ransom'].replace('Not Applicable', '-9')
            self.df['Ransom'] = self.df['Ransom'].fillna(-9).astype(float)  # Handle 'Unknown' cases

            # Handle 'Unknown' values in 'Is Hostage' before conversion
            self.df['Is Hostage'] = self.df['Is Hostage'].replace('Unknown', '-9') # Replace 'Unknown' with '-9'
            self.df['Is Hostage'] = self.df['Is Hostage'].fillna(-9).astype(int)  # Handle remaining missing values

            # Map values to descriptive strings
            self.df['Ransom'] = self.df['Ransom'].map({1: 'Yes', 0: 'No', -9: 'Unknown'})
            self.df['Is Hostage'] = self.df['Is Hostage'].map({1: 'Yes', 0: 'No', -9: 'Unknown'})
        except Exception as e:
            raise ValueError(f'Error initializing: {e}')

    def get_df(self):
        """Returns the dataframe."""
        return self.df

    def incedent_by_Years(self):
        """Returns number of incidents per year."""
        return self.df['Year'].value_counts().sort_index()

    def types_of_attack(self):
        """Returns top attack types."""
        return self.df['Attack Type 1'].value_counts().sort_values(ascending=False)

    def types_of_target(self):
        """Returns top target types."""
        return self.df['Target Type 1'].value_counts().sort_values(ascending=False)

    def top_terrorist_group(self):
        """Returns top terrorist groups."""
        return self.df['Group Name 1'].value_counts().sort_values(ascending=False)

    def avg_number_killed(self):
        """Returns average number killed."""
        return self.df['Number Killed'].mean()

    def avg_number_wounded(self):
        """Returns average number wounded."""
        return self.df['Number Wounded'].mean()

    def Incident_with_ransom(self):
        """Returns total count of incidents involving ransom demands."""
        return (self.df['Ransom'] == 'Yes').sum()

    def incidents_with_hostages(self):
        """Returns number of incidents with hostages."""
        return (self.df['Is Hostage'] == 'Yes').sum()

    def incidents_by_country(self, top_n=30):
        """Returns number of incidents per country."""
        return self.df['Country'].value_counts().head(top_n)

    def incidents_by_region(self, top_n=15):
        """Returns number of incidents per region."""
        return self.df['Region'].value_counts().head(top_n)

    def incidents_by_city(self, top_n=15):
        """Returns number of incidents per city."""
        return self.df['City'].value_counts().head(top_n)

    def incidents_by_month(self):
        """Returns number of incidents by month."""
        # Convert 'Month' column to numeric, handling non-numeric values
        self.df['Month'] = pd.to_numeric(self.df['Month'], errors='coerce')

        # Count attacks by month
        month_counts = self.df['Month'].value_counts().sort_index()

        # Ensure all months from 0 to 12 are represented, including months with zero counts
        full_month_counts = month_counts.reindex(range(0, 13), fill_value=0)

        return full_month_counts

    def incidents_by_attack_type(self):
        """Returns number of incidents per attack type."""
        return self.df['Attack Type 1'].value_counts().sort_values(ascending=False)

    def incidents_specific_attack_type(self, attack_type):
        """Returns dataframe with specific attack type."""
        return self.df[self.df['Attack Type 1'] == attack_type]

    def incidents_by_target_type(self):
        """Returns number of incidents per target type."""
        return self.df['Target Type 1'].value_counts().sort_values(ascending=False)

    def incidents_specific_target_type(self, target_type):
        """Returns dataframe with specific target type."""
        return self.df[self.df['Target Type 1'] == target_type]

    def incidents_by_year_month(self):
        """Returns number of incidents by year and month."""
        return self.df.groupby(['Year', 'Month']).size()

    def get_df_with_text_length(self):
        """Returns dataframe with text length of summaries."""
        self.df['Motive Length'] = self.df['Motive'].apply(lambda x: len(x) if pd.notna(x) else 0)
        return self.df

    def incidents_Specific_terrorist_group(self, group_name):
        """Returns dataframe with specific terrorist group."""
        return self.df[self.df['Group Name 1'] == group_name]

    def attack_by_region_sorted(self):
        """Returns number of incidents per region sorted by frequency."""
        return self.df['Region'].value_counts().sort_values(ascending=False)

    def incidents_by_region_count(self):
        """Returns count of incidents by region."""
        return self.df['Region'].value_counts()

    def frequent_regions(self, min_count):
        """Returns regions with more than min_count incidents."""
        return self.df['Region'].value_counts()[self.df['Region'].value_counts() > min_count]

    def count_attacks_by_country(self, top_n=30):
        """Returns the count of attacks by country."""
        return self.df['Country'].value_counts().head(top_n)

    def count_attacks_by_city(self, top_n=15):
        """Returns the count of attacks by city."""
        return self.df['City'].value_counts().head(top_n)

    def count_attacks_by_year(self):
        """Returns the count of attacks by year."""
        return self.df['Year'].value_counts().sort_index()

    def count_attacks_by_month(self):
        """Returns the count of attacks by month, sorted in ascending order."""
        # Convert 'Month' column to numeric, handling non-numeric values
        self.df['Month'] = pd.to_numeric(self.df['Month'], errors='coerce')

        # Count attacks by month
        month_counts = self.df['Month'].value_counts().sort_index()

        # Ensure all months from 0 to 12 are represented, including months with zero counts
        full_month_counts = month_counts.reindex(range(0, 13), fill_value=0)

        return full_month_counts

    def count_attack_types(self, top_n=15):
        """Returns the count of attack types."""
        return self.df['Attack Type 1'].value_counts().head(top_n)

    def count_target_types(self, top_n=15):
        """Returns the count of target types."""
        return self.df['Target Type 1'].value_counts().head(top_n)

    def count_terrorist_groups(self, top_n=15):
        """Returns the count of terrorist groups."""
        return self.df['Group Name 1'].value_counts().head(top_n)

    def total_fatalities(self):
        """Returns the total number of fatalities."""
        self.df['Number Killed'] = pd.to_numeric(self.df['Number Killed'], errors='coerce')
        return self.df['Number Killed'].sum()

    def total_injuries(self):
        """Returns the total number of injuries."""
        self.df['Number Wounded'] = pd.to_numeric(self.df['Number Wounded'], errors='coerce')
        return self.df['Number Wounded'].sum()

    def avg_fatalities_per_attack(self):
        """Returns the average number of fatalities per attack."""
        return self.df['Number Killed'].mean()

    def avg_injuries_per_attack(self):
        """Returns the average number of injuries per attack."""
        return self.df['Number Wounded'].mean()

    def count_incidents_with_hostages(self):
        """Returns the count of incidents with hostages."""
        return (self.df['Is Hostage'] == 'Yes').sum()

    def count_incidents_with_ransom(self):
        """Returns the count of incidents with ransom demands."""
        return (self.df['Ransom'] == 'Yes').sum()

    def count_unknown_groups_per_year(self):
        """Returns the count of 'Unknown' group names per year in descending order."""
        try:
            unknown_counts = self.df[self.df['Group Name 1'] == 'Unknown'].groupby('Year').size()
            return unknown_counts.sort_values(ascending=False)
        except Exception as e:
            raise ValueError(f'Error occurred: {e}')

    def percentage_unknown_groups_per_year(self):
        """Returns the percentage of unknown group attacks by year."""
        try:
            # Count unknown group attacks per year
            unknown_counts = self.df[self.df['Group Name 1'] == 'Unknown'].groupby('Year').size()

            # Count total attacks per year
            total_attacks = self.df.groupby('Year').size()

            # Calculate the percentage of unknown group attacks
            percentage_unknown = (unknown_counts / total_attacks * 100).fillna(0)

            return percentage_unknown.sort_index()  # Return in ascending order by year
        except Exception as e:
            raise ValueError(f'Error occurred: {e}')

    def active_years_of_top_20_groups(self):
        """Returns the top 20 terrorist groups with the highest number of incidents and their number of active years."""
        try:
            # Filter out 'Unknown' groups
            known_groups_df = self.df[self.df['Group Name 1'] != 'Unknown']

            # Count incidents per terrorist group
            group_incident_counts = known_groups_df['Group Name 1'].value_counts()

            # Get the top 20 groups by number of incidents
            top_20_groups = group_incident_counts.head(20).index

            # Count active years for each of these top 20 groups
            active_years = known_groups_df[known_groups_df['Group Name 1'].isin(top_20_groups)].groupby('Group Name 1')['Year'].nunique()

            # Combine the incident counts and active years into a DataFrame for better readability
            top_20_summary = pd.DataFrame({
                'Number of Incidents': group_incident_counts.loc[top_20_groups],
                'Active Years': active_years
            }).sort_values(by='Number of Incidents', ascending=False)

            # Return only the 'Active Years' for top 20 groups
            return top_20_summary[['Active Years']]

        except Exception as e:
            raise ValueError(f'Error occurred: {e}')


    def highest_groups_logistical_status(self):
        """Find the groups with the highest number of incidents where Logistical Status is 1."""
        try:
            # Filter the DataFrame for Logistical Status 1
            logistical_df = self.df[self.df['Logistical Status'] == 1]

            # Count incidents per terrorist group
            group_incident_counts = logistical_df['Group Name 1'].value_counts()

            return group_incident_counts

        except Exception as e:
            raise ValueError(f'Error occurred: {e}')

    def attacks_by_group_ideological_status(self):
        """Find the number of attacks for each group with Ideological Status as 1."""
        try:
            # Filter the DataFrame for Ideological Status 1
            ideological_df = self.df[self.df['Ideological Status'] == 1]

            # Group by 'Group Name 1' and count the number of attacks
            attacks_by_group = ideological_df.groupby('Group Name 1').size()

            # Sort the results to find the groups with the highest number of attacks
            sorted_attacks = attacks_by_group.sort_values(ascending=False)

            return sorted_attacks

        except Exception as e:
            raise ValueError(f'Error occurred: {e}')

    def casualties_by_weapon_type(self, region):
        """Returns count of casualties by weapon type, region-wise."""
        try:
            region_df = self.df.loc[self.df['Region'] == region]
            # Convert 'Number Killed' to numeric, handling errors
            region_df.loc[:, 'Number Killed'] = pd.to_numeric(region_df['Number Killed'], errors='coerce')
            # Group by weapon type and sum the casualties
            casualties_by_weapon = region_df.groupby('Weapon Type 1')['Number Killed'].sum().reset_index()
            casualties_by_weapon.rename(columns={'Number Killed': 'Total Casualties'}, inplace=True)
            return casualties_by_weapon.sort_values(by='Total Casualties', ascending=False)
        except Exception as e:
            raise ValueError(f'Error occurred: {e}')


gtd = TerrorismWrangling(gtd_df)

#Usage Example:

#gtd.Incident_with_ransom
#gtd.incidents_with_hostages()
#gtd.incidents_by_country()
#gtd.incidents_by_region()
#gtd.incidents_by_city()
#gtd.incidents_by_month()
#gtd.incidents_by_attack_type()
#gtd.incidents_by_target_type()
#gtd.incidents_by_year_month()
#gtd.get_df_with_text_length()
#gtd.incidents_Specific_terrorist_group('Taliban')
#gtd.attack_by_region_sorted()
#gtd.incidents_by_region_count()
#gtd.frequent_regions(100)
#gtd.count_attacks_by_country()
#gtd.count_attacks_by_city()
#gtd.count_attacks_by_year()
#gtd.count_attacks_by_month()
#gtd.count_attack_types()
#gtd.count_target_types()
#gtd.count_terrorist_groups()
#gtd.total_fatalities()
#gtd.total_injuries()
#gtd.avg_fatalities_per_attack()
#gtd.avg_injuries_per_attack()
#gtd.count_incidents_with_hostages()
#gtd.count_incidents_with_ransom()
#gtd.count_unknown_groups_per_year()
#gtd.percentage_unknown_groups_per_year()
#gtd.active_years_of_top_20_groups()
#gtd.highest_groups_logistical_status()
#gtd.attacks_by_group_ideological_status()
#gtd.casualties_by_weapon_type('Middle East & North Africa')
#gtd.get_df()
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

````python
#Class for Visualization.

class TerrorismVisualization:
    def __init__(self, df):
        """Initialize with DataFrame."""
        self.df = df

    def incidents_by_year(self):
        """Plot number of incidents year-wise."""
        try:
            data = self.df.groupby('Year').size()

            fig = go.Figure()
            fig.add_trace(go.Scatter(
                x=data.index,
                y=data.values,
                mode='lines+markers',
                marker=dict(color='black', size=10),
                line=dict(color='black'),
                text=[f'Year: {year}<br>Incidents: {num}' for year, num in zip(data.index, data.values)],
                hoverinfo='text'
            ))

            fig.update_layout(
                title='Number of Incidents Year wise',
                xaxis_title='Year',
                yaxis_title='Number of Incidents',
                xaxis=dict(tickangle=45),
                plot_bgcolor='slategrey',
                paper_bgcolor='lightsteelblue',
                template='plotly_white'
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")

    def attacks_by_country(self):
        """Plot map of total number of attacks by country."""
        try:
            gtd_map_df = self.df[['Country', 'Year']]
            terr = gtd_map_df.groupby('Country', as_index=False).size()
            terr = terr.rename(columns={'size': 'Attacks'})

            fig = px.choropleth(
                terr,
                locations='Country',
                locationmode='country names',
                color='Attacks',
                hover_name='Country',
                projection='orthographic',
                title='Total number of attacks (1970-2017)-Country Wise',
                labels={'Attacks': 'Number of Attacks'}
            )

            fig.update_layout(
                autosize=False,
                width=1200,
                height=800
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")

    def attacks_by_region(self):
        """Plot bar chart of attacks by region."""
        try:
            region_counts = self.df['Region'].value_counts().reset_index()
            region_counts.columns = ['Region', 'Number of Attacks']

            fig = px.bar(
                region_counts,
                x='Region',
                y='Number of Attacks',
                color='Number of Attacks',
                color_continuous_scale=px.colors.sequential.Plasma,
                title='Attacks by Region',
                labels={'Number of Attacks': 'Number of Attacks'},
                text='Number of Attacks'
            )

            fig.update_layout(
                xaxis_title='Region',
                yaxis_title='Number of Attacks',
                xaxis_tickangle=-45,
                coloraxis_colorbar_title='Number of Attacks'
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")

    def attack_types(self):
        """Plot pie chart for attack types (Type 1)."""
        try:
            attack_type_counts = self.df['Attack Type 1'].value_counts().reset_index()
            attack_type_counts.columns = ['Attack Type 1', 'Number of Attacks']

            fig = px.pie(
                attack_type_counts,
                names='Attack Type 1',
                values='Number of Attacks',
                title='Types of Attack. Type-1',
                labels={'Number of Attacks': 'Number of Attacks'},
                template='plotly_white'
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")


    def attack_target_heatmap(self):
        """Plot and save a heatmap of correlation between attack types and target types."""
        try:
            # Generate the crosstab
            attack_target_crosstab = pd.crosstab(self.df['Attack Type 1'], self.df['Target Type 1'])

            # Define colormap similar to Plotly's Plasma
            cmap = plt.get_cmap('plasma')

            # Set up the figure
            plt.figure(figsize=(16, 6), dpi=150)
            sns.heatmap(
                attack_target_crosstab,
                annot=True,
                cmap=cmap,
                linewidths=.5,
                fmt="d",
                annot_kws={"size": 10}  # Adjust font size for annotations
            )
            plt.title('Correlation Heatmap Between Attack Types and Target Types')
            plt.ylabel('Attack Type')
            plt.xlabel('Target Type')
            plt.xticks(rotation=45, ha='right')

            # Save the chart as an image
            plt.savefig('attack_target_heatmap.png', bbox_inches='tight', dpi=150)

            # Show the chart
            plt.show()

        except Exception as e:
            print(f"Error occurred: {e}")


    def attacks_by_group(self):
        """Plot bar chart of number of attacks by the top 20 terrorist groups."""
        try:
            # Get the top 20 groups based on the number of attacks
            top_20_groups = self.df['Group Name 1'].value_counts().head(20).reset_index()
            top_20_groups.columns = ['Group Name 1', 'Number of Attacks']

            fig = px.bar(
                top_20_groups,
                x='Group Name 1',
                y='Number of Attacks',
                color='Number of Attacks',
                color_continuous_scale=px.colors.sequential.Plasma,
                title='Top 20 Terrorist Groups by Number of Attacks',
                labels={'Number of Attacks': 'Number of Attacks'},
                text='Number of Attacks'
            )

            fig.update_layout(
                xaxis_title='Group Name 1',
                yaxis_title='Number of Attacks',
                xaxis_tickangle=-45,
                coloraxis_colorbar_title='Number of Attacks'
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")

    def incidents_and_unknown_attackers_by_year(self):
        """Plot incidents and unknown attackers by year with overlapping bars."""
        try:
            # Get data for incidents and unknown attackers
            attacks_per_year = self.df.groupby('Year').size()
            unknown_counts_per_year = self.df[self.df['Group Name 1'] == 'Unknown'].groupby('Year').size()
            unknown_counts_per_year = unknown_counts_per_year.reindex(attacks_per_year.index, fill_value=0)
            percentages = (unknown_counts_per_year / attacks_per_year * 100).fillna(0)

            # Create the figure
            fig = go.Figure()

            # Add bar chart for total incidents
            fig.add_trace(go.Bar(
                x=attacks_per_year.index,
                y=attacks_per_year.values,
                name='Total Incidents',
                marker_color='black',
                opacity=0.7,
                text=[f'Year: {year}<br>Total Incidents: {count}' for year, count in zip(attacks_per_year.index, attacks_per_year.values)],
                texttemplate='%{text}',
                textposition='inside'
            ))

            # Add overlapping bar chart for unknown attackers count
            fig.add_trace(go.Bar(
                x=unknown_counts_per_year.index,
                y=unknown_counts_per_year.values,
                name='Unknown Attackers Count',
                marker_color='red',
                opacity=0.7,
                text=[f'Year: {year}<br>Incidents with unknown attackers: {unknown}<br>Percentage: {perc:.1f}%' for year, unknown, perc in zip(unknown_counts_per_year.index, unknown_counts_per_year.values, percentages)],
                texttemplate='%{text}',
                textposition='inside'
            ))

            # Update layout
            fig.update_layout(
                title='Number of Incidents and Unknown Attackers per Year',
                xaxis_title='Year',
                yaxis_title='Number of Incidents',
                barmode='overlay',  # Overlay bars on top of each other
                yaxis=dict(range=[0, max(attacks_per_year.max(), unknown_counts_per_year.max()) * 1.1]),  # Adjust range as needed
                legend_title='Legend'
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")


    def active_years_for_top_20_groups(self):
        """Plot active years for the top 20 terrorist groups based on the number of attacks, with hover text showing first and last attack years and total number of attacks."""
        try:
            # Filter out 'Unknown'
            filtered_df = self.df[self.df['Group Name 1'] != 'Unknown']

            # Get the top 20 groups based on the number of attacks
            top_20_groups = filtered_df['Group Name 1'].value_counts().head(20)
            top_20_group_names = top_20_groups.index

            # Filter the DataFrame to only include rows from the top 20 groups
            top_20_df = filtered_df[filtered_df['Group Name 1'].isin(top_20_group_names)]

            # Calculate the first and last attack years for each group
            group_years = top_20_df.groupby('Group Name 1')['Year'].agg(['min', 'max'])
            group_years['Active Years'] = group_years['max'] - group_years['min'] + 1
            group_years['Number of Attacks'] = top_20_groups  # Include the count of attacks for each group

            # Ensure order matches the top 20 by number of attacks
            group_years = group_years.loc[top_20_group_names]

            # Generate hover text with first and last attack years, active years, and number of attacks
            hover_text = [
                f"Group: {group}<br>First Incident Year: {row['min']}<br>Last Incident Year: {row['max']}<br>Active Years: {row['Active Years']}<br>Number of Attacks: {row['Number of Attacks']}"
                for group, row in group_years.iterrows()
            ]

            # Plot the chart
            fig = px.bar(
                group_years.reset_index(),
                x='Group Name 1',
                y='Active Years',
                color='Active Years',
                color_continuous_scale=px.colors.sequential.Viridis,
                title='Active Years of Top 20 Terrorist Groups by Number of Attacks',
                labels={'Group Name 1': 'Group Name', 'Active Years': 'Number of Active Years'},
                text='Active Years',
                hover_name=hover_text  # Custom hover text
            )

            # Adjust layout for better readability
            fig.update_layout(
                xaxis_title='Group Name',
                yaxis_title='Number of Active Years',
                xaxis_tickangle=-45,
                coloraxis_colorbar_title='Active Years',
                plot_bgcolor='white',
                width=900,  # Adjusted for better visibility
                height=600
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")


    def number_killed_by_region_by_weapon(self, value_col, category_col, animation_col):
        try:
            # Ensure the specified value column is numeric and handle missing values
            self.df.loc[:, value_col] = pd.to_numeric(self.df[value_col], errors='coerce')
            self.df = self.df.dropna(subset=[value_col])

            # Group by the specified category and animation columns, summing the values
            grouped_df = self.df.groupby([category_col, animation_col])[value_col].sum().reset_index()

            # Initialize figure with the first frame of the pie chart
            first_animation_value = grouped_df[animation_col].iloc[0]
            first_frame = grouped_df[grouped_df[animation_col] == first_animation_value]
            fig = go.Figure(data=[
                go.Pie(
                    labels=first_frame[category_col],
                    values=first_frame[value_col],
                    name=str(first_animation_value)
                )
            ])

            # Add frames for animation for each unique value in the animation column
            frames = [
                go.Frame(
                    data=[go.Pie(
                        labels=grouped_df.loc[grouped_df[animation_col] == anim_value, category_col],
                        values=grouped_df.loc[grouped_df[animation_col] == anim_value, value_col]
                    )],
                    name=str(anim_value)
                )
                for anim_value in grouped_df[animation_col].unique()
            ]

            fig.frames = frames

            # Update layout with play and pause buttons, and animation settings
            fig.update_layout(
                title='Distribution of Number Killed by Region and Weapon Type',
                updatemenus=[{
                    'buttons': [
                        {
                            'args': [None, {'frame': {'duration': 1000, 'redraw': True}, 'fromcurrent': True}],
                            'label': 'Play',
                            'method': 'animate'
                        },
                        {
                            'args': [[None], {'frame': {'duration': 0, 'redraw': True}, 'mode': 'immediate',
                                              'transition': {'duration': 0}}],
                            'label': 'Pause',
                            'method': 'animate'
                        }
                    ],
                    'direction': 'left',
                    'pad': {'r': 10, 't': 87},
                    'showactive': True,
                    'type': 'buttons',
                    'x': 0.1,
                    'xanchor': 'right',
                    'y': 0,
                    'yanchor': 'top'
                }],
                sliders=[{
                    'steps': [
                        {
                            'label': str(anim_value),
                            'method': 'animate',
                            'args': [
                                [str(anim_value)],
                                {'mode': 'immediate', 'frame': {'duration': 1000, 'redraw': True},
                                 'transition': {'duration': 300}}
                            ]
                        }
                        for anim_value in grouped_df[animation_col].unique()
                    ],
                    'currentvalue': {'prefix': f'{animation_col}: '}
                }]
            )

            fig.show()
        except Exception as e:
            print(f"Error occurred: {e}")

    def attack_count_by_region_by_weapon_by_group_by_target_type(self):
        """Create a scatter plot showing attack count by region, weapon, target type, and group."""
        count_df = self.df.query("Year >= 1970 and Year <= 2017") \
            .groupby(['Region', 'Target Type 1', 'Group Name 1', 'Weapon Type 1']) \
            .size() \
            .reset_index(name='Attack Count')

        # Create the scatter plot
        fig = px.scatter(
            count_df,
            title="Attack Count by Region by Weapon by Target Type",
            x="Attack Count",  # Total count of attacks on each target type
            y="Region",
            color="Target Type 1",
            hover_name="Group Name 1",
            animation_frame="Weapon Type 1",
            template="plotly_dark",
            log_x=True
        )
        fig.show()

gtd_viz = TerrorismVisualization(gtd_df)

#Usage Examples

#CHART-1- gtd_viz.incidents_by_year()
#CHART-2- gtd_viz.attacks_by_country()
#CHART-3- gtd_viz.attacks_by_region()
#CHART-4- gtd_viz.attack_types()
#CHART-5- gtd_viz.attack_target_heatmap()
#CHART-6- gtd_viz.attacks_by_group()
#CHART-7- gtd_viz.incidents_and_unknown_attackers_by_year()
#CHART-8- gtd_viz.active_years_for_top_20_groups()
#CHART-9- gtd_viz.number_killed_by_region_by_weapon()
#CHART-10- gtd_viz.attack_count_by_region_by_weapon_by_group_bytarget_type()
````

## **Chart - 5** Heatmap Between Attack Types and Target Types

````python
#Chart - 5
#Correlation Heatmap Between Attack Types and Target Types
gtd_viz.attack_target_heatmap()
````

##### 1. Why did you pick the specific chart?

1. Identifying Relationships: A heatmap can help visualize the correlation between categorical variables by converting them into numerical representations. This allows you to see how closely related different attack types are to various target types.
